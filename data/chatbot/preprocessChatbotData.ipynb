```python
# 공원/쇼핑 데이터 불러오기
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/capstone/chatbotData/1. 공원, 쇼핑(문화).csv')

data
```

```python
# 공원/쇼핑 MLSFC_NM/FCLTY_NM/LNM_ADDR 컬럼 추출, 컬럼 순서 및 컬럼명 변경
data = data.filter(['MLSFC_NM', 'FCLTY_NM', 'LNM_ADDR'])

data = data[['FCLTY_NM', 'MLSFC_NM', 'LNM_ADDR']]

data.columns = ['name', 'type', 'address']

data
```

```python
# 공원/쇼핑 데이터 주소 지역명 통합하기
def replace_address_prefix(address):
  # 주소를 공백으로 분할
  parts = address.split()
  # 분할된 첫 번째 단어에 따라 지역명을 통합한 이름으로 치환
  regions = {
    '서울': '서울특별시', '부산': '부산광역시', '대구': '대구광역시',
    '인천': '인천광역시', '광주': '광주광역시', '대전': '대전광역시',
    '울산': '울산광역시', '세종': '세종특별자치시', '경기': '경기도',
    '강원': '강원도', '충북': '충청북도', '충남': '충청남도',
    '전북': '전라북도', '전남': '전라남도', '경북': '경상북도',
    '경남': '경상남도'
  }
  if parts[0] in regions:
    parts[0] = regions[parts[0]]
  return ' '.join(parts)

data['address'] = data['address'].apply(replace_address_prefix)

data
```

```python
# 공원/쇼핑 데이터 분할하기
park = data[data['type'] == '자연_공원']
shopping = data[data['type'] == '쇼핑시설_전통시장']

print(park)
print(shopping)
```

```python
# 교육 데이터에 전화번호 컬럼 생성 후 해당값 google places api 활용하여 추가하기
import requests

# 전화번호 가져오기 함수
def get_phone_number(place_name):
  # 장소 검색으로 place_id 찾기
  search_params = {
    'input': place_name,
    'inputtype': 'textquery',
    'fields': 'place_id',
    'key': api_key
  }
  search_response = requests.get(search_url, params=search_params)
  search_results = search_response.json()

  # 검색 결과가 없으면 '정보 없음' 반환
  if not search_results.get('candidates'):
    return '정보 없음'

  place_id = search_results['candidates'][0].get('place_id')
  if not place_id:
    return '정보 없음'  # 장소 ID를 찾지 못한 경우

  # place_id로 전화번호 가져오기
  details_params = {
    'place_id': place_id,
    'fields': 'formatted_phone_number',
    'key': api_key
  }
  details_response = requests.get(details_url, params=details_params)
  details_results = details_response.json()

  # 전화번호가 있으면 반환, 없으면 '정보 없음'
  return details_results['result'].get('formatted_phone_number', '정보 없음')

# 전화번호 컬럼 추가 및 값 채우기
api_key = 'api_key'
search_url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
details_url = 'https://maps.googleapis.com/maps/api/place/details/json'

# 'phoneNumber' 컬럼 추가
for index, row in education.iterrows():
  education.at[index, 'phoneNumber'] = get_phone_number(row['name'])

education
```

```python
# 간호, 목욕, 요양 데이터 pdf에서 테이블 추출하여 데이터 프레임으로 만들기
import pdfplumber
import pandas as pd

# pdf 테이블 추출 및 데이터 프레임 변환 함수
def pdf_table_to_dataframe(pdf_path, start_page, end_page):
  all_tables = []  # 모든 추출된 테이블을 저장할 리스트
  region_name = ""  # 현재 지역 이름을 저장할 변수

  with pdfplumber.open(pdf_path) as pdf:
    for i in range(start_page, end_page + 1):
      page = pdf.pages[i]
      text = page.extract_text()
      tables = page.extract_tables()

      # 지역 이름을 찾아 리스트에 저장
      region_names = [line for line in text.split('\n') if "▣" in line]
      # 다음 지역 이름과 관련된 테이블을 찾기 위한 인덱스
      next_region_index = 0

      for table in tables:
        if next_region_index < len(region_names):
          region_name = region_names[next_region_index].strip("▣ ").strip()
          next_region_index += 1

        # 테이블 데이터를 데이터 프레임으로 변환
        df = pd.DataFrame(table[1:], columns=table[0])

        # 시군구 컬럼의 값이 None인 행, '합계'가 포함된 행 제거 및 컬럼명이 None인 열 제거
        df.dropna(subset=['시･군･구'], inplace=True)
        df = df[~df['시･군･구'].str.contains('합계', na=False)]
        df = df.loc[:, df.columns.notnull()]

        # 소재지 컬럼 확인
        if '소 재 지' in df.columns:
          df['소 재 지'] = region_name + " " + df['소 재 지'].astype(str)

        # 데이터 처리 후 최종 리스트에 추가
        all_tables.append(df)

  if all_tables:
    final_df = pd.concat(all_tables, ignore_index=True)
    return final_df
  else:
    return pd.DataFrame()  # 추출된 테이블이 없는 경우 빈 DataFrame 반환

care = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/chatbotData/2. 방문간호서비스(방문서비스-간호).pdf', 1, 15)
bath = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/chatbotData/2. 방문목욕서비스(방문서비스-목욕)-536.pdf', 1, 159)
recuperation1 = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/chatbotData/2. 방문요양서비스(방문서비스-요양).pdf', 1, 270)
recuperation2 = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/chatbotData/2. 재가노인지원서비스(방문서비스-요양).pdf', 1, 19)
```

```python
# 간호, 목욕, 요양 데이터 시설명, 소재지, 전화번호만 추출하기
care = care.filter(['시 설 명', '소 재 지', '전화번호\n(FAX번호)'])
bath = bath.filter(['시 설 명', '소 재 지', '전화번호\n(FAX번호)'])
recuperation = recuperation.filter(['시 설 명', '소 재 지', '전화번호\n(FAX번호)'])

print(care)
print(bath)
print(recuperation)
```

```python
# 간호, 목욕, 요양 데이터 시설명 컬럼 '\n' -> ' ' 대체하기
care['시 설 명'] = care['시 설 명'].str.replace('\n', '')
bath['시 설 명'] = bath['시 설 명'].str.replace('\n', '')
recuperation['시 설 명'] = recuperation['시 설 명'].str.replace('\n', '')

# 간호, 목욕, 요양 데이터 소재지 컬럼 '\n' -> ' ' 대체하기
care['소 재 지'] = care['소 재 지'].str.replace('\n', ' ')
bath['소 재 지'] = bath['소 재 지'].str.replace('\n', ' ')
recuperation['소 재 지'] = recuperation['소 재 지'].str.replace('\n', ' ')

print(care)
print(bath)
print(recuperation)
```

```python
# 간호, 목욕, 요양 데이터 전화번호\n(FAX번호) 컬럼 -> '\n' 포함 뒷 부분 제거
care['전화번호\n(FAX번호)'] = care['전화번호\n(FAX번호)'].str.split('\n').str[0]
bath['전화번호\n(FAX번호)'] = bath['전화번호\n(FAX번호)'].str.split('\n').str[0]
recuperation['전화번호\n(FAX번호)'] = recuperation['전화번호\n(FAX번호)'].str.split('\n').str[0]

print(care)
print(bath)
print(recuperation)
```

```python
# 간호, 목욕, 요양 데이터 컬럼명 영문으로 변경, 시설 타입(type) 컬럼 추가하기
care.columns = ['name', 'address', 'phoneNumber']
bath.columns = ['name', 'address', 'phoneNumber']
recuperation.columns = ['name', 'address', 'phoneNumber']

care.insert(1, 'type', 'visit_care')
bath.insert(1, 'type', 'visit_bath')
recuperation.insert(1, 'type', 'visit_recuperation')

print(care)
print(bath)
print(recuperation)
```

```python
# 간호, 목욕, 요양 데이터 주소 기준 오름차순으로 정럴하기
care = care.sort_values(by='address', ignore_index=True)
bath = bath.sort_values(by='address', ignore_index=True)
recuperation = recuperation.sort_values(by='address', ignore_index=True)

print(care)
print(bath)
print(recuperation)
```
