# 병원, 약국 데이터 전처리

```python
# 병원, 약국 데이터 불러오기
import pandas as pd
hospital = pd.read_excel('/content/drive/MyDrive/capstone/mapData/1. 병원(병원:약국).xlsx')
pharmacy = pd.read_excel('/content/drive/MyDrive/capstone/mapData/1. 약국(병원:약국).xlsx')

print(hospital)
print(pharmacy)
```

```python
# 병원, 약국 데이터 기관명, 주소, 전화번호, 좌표(x), 좌표(y) 추출하기
hospital = hospital.filter(['요양기관명', '주소', '전화번호', '좌표(X)', '좌표(Y)'])
pharmacy = pharmacy.filter(['요양기관명', '주소', '전화번호', '좌표(x)', '좌표(y)'])

print(hospital)
print(pharmacy)
```

```python
# 병원, 약국 데이터 컬럼 순서 변경, 컬럼명 영문으로 변경, 시설 타입(type) 컬럼 추가하기
hospital = hospital[['요양기관명', '주소', '전화번호', '좌표(Y)', '좌표(X)']]
pharmacy = pharmacy[['요양기관명', '주소', '전화번호', '좌표(y)', '좌표(x)']]

hospital.columns = ['name', 'address', 'phoneNumber', 'latitude', 'longitude']
pharmacy.columns = ['name', 'address', 'phoneNumber', 'latitude', 'longitude']

hospital.insert(1, 'type', 'hospital')
pharmacy.insert(1, 'type', 'pharmacy')

print(hospital)
print(pharmacy)
```

```python
# 병원, 약국 데이터 결측값 확인하기
print(hospital.info())
print(pharmacy.info())
```

```python
# 병원, 약국 데이터 전화번호 '-' 제거, 전화번호 결측값 '' 대체하기
hospital['phoneNumber'] = hospital['phoneNumber'].str.replace('-', '').fillna('')
pharmacy['phoneNumber'] = pharmacy['phoneNumber'].str.replace('-', '').fillna('')

print(hospital)
print(pharmacy)
```

```python
# 병원, 약국 데이터 위경도 결측값 naver cloud platform geocode api로 채우기
import numpy as np
from urllib.request import urlopen
from urllib import parse
from urllib.request import Request
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import json

# 위경도 결측값 처리하기 함수
def geo_latlng(address):
  add_urlenc = parse.quote(address)
  url = api_url + add_urlenc
  request = Request(url)
  request.add_header('X-NCP-APIGW-API-KEY-ID', client_id)
  request.add_header('X-NCP-APIGW-API-KEY', client_secret)
  try:
    response = urlopen(request)
    rescode = response.getcode()
    if rescode == 200:
      response_body = response.read().decode('utf-8')
      response_body = json.loads(response_body)
      if response_body['addresses']:
        latitude = response_body['addresses'][0]['y']
        longitude = response_body['addresses'][0]['x']
        print("Success!")
        return latitude, longitude
      else:
        print("No result!")
        return np.nan, np.nan
    else:
      print('Response error code:', rescode)
      return np.nan, np.nan
  except HTTPError as e:
    print('HTTP Error!')
    return np.nan, np.nan

# 위경도 결측값 처리하기
client_id = 'client_id'
client_secret = 'client_secret'

api_url = 'https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query='

# 병원 데이터
for idx, row in hospital.iterrows():
  if pd.isnull(row['latitude']) or pd.isnull(row['longitude']):
    lat, lng = geo_latlng(row['address'])
    hospital.at[idx, 'latitude'] = lat
    hospital.at[idx, 'longitude'] = lng

# 약국 데이터
for idx, row in pharmacy.iterrows():
  if pd.isnull(row['latitude']) or pd.isnull(row['longitude']):
    lat, lng = geo_latlng(row['address'])
    pharmacy.at[idx, 'latitude'] = lat
    pharmacy.at[idx, 'longitude'] = lng


print(hospital)
print(pharmacy)
```

```python
# 복지주택, 복지관, 케어센터 데이터 pdf에서 테이블 추출하여 데이터 프레임으로 만들기
import pdfplumber
import pandas as pd

# pdf 테이블 추출 및 데이터 프레임 변환 함수
def pdf_table_to_dataframe(pdf_path, start_page, end_page):
  all_tables = []  # 모든 추출된 테이블을 저장할 리스트
  region_name = ""  # 현재 지역 이름을 저장할 변수

  with pdfplumber.open(pdf_path) as pdf:
    for i in range(start_page, end_page + 1):
      page = pdf.pages[i]
      text = page.extract_text()
      tables = page.extract_tables()

      # 지역 이름을 찾아 리스트에 저장
      region_names = [line for line in text.split('\n') if "▣" in line]
      # 다음 지역 이름과 관련된 테이블을 찾기 위한 인덱스
      next_region_index = 0

      for table in tables:
        if next_region_index < len(region_names):
          region_name = region_names[next_region_index].strip("▣ ").strip()
          next_region_index += 1

        # 테이블 데이터를 데이터 프레임으로 변환
        df = pd.DataFrame(table[1:], columns=table[0])

        # 시군구 컬럼의 값이 None인 행, '합계'가 포함된 행 제거 및 컬럼명이 None인 열 제거
        df.dropna(subset=['시･군･구'], inplace=True)
        df = df[~df['시･군･구'].str.contains('합계', na=False)]
        df = df.loc[:, df.columns.notnull()]

        # 소재지 컬럼 확인
        if '소재지' in df.columns:
          df['소재지'] = region_name + " " + df['소재지'].astype(str)
        elif '소 재 지' in df.columns:
          df['소 재 지'] = region_name + " " + df['소 재 지'].astype(str)

        # 데이터 처리 후 최종 리스트에 추가
        all_tables.append(df)

  if all_tables:
    final_df = pd.concat(all_tables, ignore_index=True)
    return final_df
  else:
    return pd.DataFrame()  # 추출된 테이블이 없는 경우 빈 DataFrame 반환

welfarehouse = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/mapData/2. 복지주택.pdf', 1, 3)
welfarecenter = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/mapData/3. 복지관.pdf', 3, 21)
carecenter1 = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/mapData/4. 단기보호서비스(케어센터).pdf', 3, 8)
carecenter2 = pdf_table_to_dataframe('/content/drive/MyDrive/capstone/mapData/4. 주야간보호서비스(케어센터).pdf', 1, 143)
```

```python
# 복지주택, 복지관, 케어센터 데이터 시설명, 소재지, 전화번호만 추출하기
welfarehouse = welfarehouse.filter(['시설명', '소재지', '전화번호\n(FAX번호)'])
welfarecenter = welfarecenter.filter(['시 설 명', '소 재 지', '전화번호\n(FAX번호)'])
carecenter = carecenter.filter(['시 설 명', '소 재 지', '전화번호\n(FAX번호)'])

print(welfarehouse)
print(welfarecenter)
print(carecenter)
```

```python
# 복지주택, 복지관, 케어센터 데이터 시설명 컬럼 '\n' -> ' ' 대체하기
welfarehouse['시설명'] = welfarehouse['시설명'].str.replace('\n', ' ')
welfarecenter['시 설 명'] = welfarecenter['시 설 명'].str.replace('\n', ' ')
carecenter['시 설 명'] = carecenter['시 설 명'].str.replace('\n', ' ')

# 복지주택, 복지관, 케어센터 데이터 소재지 컬럼 '\n' -> ' ' 대체하기
welfarehouse['소재지'] = welfarehouse['소재지'].str.replace('\n', ' ')
welfarecenter['소 재 지'] = welfarecenter['소 재 지'].str.replace('\n', ' ')
carecenter['소 재 지'] = carecenter['소 재 지'].str.replace('\n', ' ')

print(welfarehouse)
print(welfarecenter)
print(carecenter)
```

```python
# 복지주택, 복지관, 케어센터 데이터 전화번호\n(FAX번호) 컬럼 -> '\n' 포함 뒷 부분 제거, '-' 제거
welfarehouse['전화번호\n(FAX번호)'] = welfarehouse['전화번호\n(FAX번호)'].str.split('\n').str[0].str.replace('-', '')
welfarecenter['전화번호\n(FAX번호)'] = welfarecenter['전화번호\n(FAX번호)'].str.split('\n').str[0].str.replace('-', '')
carecenter['전화번호\n(FAX번호)'] = carecenter['전화번호\n(FAX번호)'].str.split('\n').str[0].str.replace('-', '')

print(welfarehouse)
print(welfarecenter)
print(carecenter)
```

```python
# 복지주택, 복지관, 케어센터 데이터 컬럼명 영문으로 변경, 시설 타입(type) 컬럼 추가하기
welfarehouse.columns = ['name', 'address', 'phoneNumber']
welfarecenter.columns = ['name', 'address', 'phoneNumber']
carecenter.columns = ['name', 'address', 'phoneNumber']

welfarehouse.insert(1, 'type', 'welfarehouse')
welfarecenter.insert(1, 'type', 'welfarecenter')
carecenter.insert(1, 'type', 'carecenter')

print(welfarehouse)
print(welfarecenter)
print(carecenter)
```
